```python
Y = df["price"]
features = df.drop(columns='price')
```

```python
numerical_features = features._get_numeric_data().columns 
categorical_features = features.columns.difference(numerical_features)
```

```python
num_data = df[numerical_features]
cat_data = df[categorical_features] 
```

```python
#Encoding Categorical Variable
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cat_data = cat_data.astype(str).apply(LabelEncoder().fit_transform)
```

```python
#Normalizing Numerical Variable
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

ss =StandardScaler()
    
for col in num_data.columns:
    num_data[col] = ss.fit_transform(num_data[[col]]).squeeze()
```

```python
train_data = pd.concat([num_data, cat_data], axis = 1)
```

```python
data = pd.concat([train_data, Y], axis = 1)
```

```python

Correlation_Matrix = data.corr()
Most_Correlated_Features = Correlation_Matrix.index[abs(Correlation_Matrix["price"])>0.2]
sns.heatmap(df[Most_Correlated_Features].corr(),annot=True)
```

```python
Most_Correlated_Features
```

```python
model_data = data[['wheelbase', 'carlength', 'carwidth', 'curbweight', 'enginesize',
       'boreratio', 'horsepower', 'citympg', 'highwaympg', 'CarName',
       'drivewheel', 'enginelocation', 'fuelsystem', 'price']]
```

```python
model_data.head()
```

```python
#Declaring features and target
Y = model_data["price"]
X = model_data.drop(columns='price')
```

```python
# Split the data into training and test samples
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#Building Linear regression model
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
#fitting model to training dataset
linear_model.fit(X_train, Y_train)

#obtaining predictions
predicted_values = linear_model.predict(X_test)
```

```python
linear_model.score(X_test, Y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf =  RandomForestRegressor(random_state=42)
rf = rf.fit(X_train, Y_train)
```

```python
rf.score(X_test, Y_test)
```

```python
feature_imp = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)

ax = feature_imp.plot(kind='bar', figsize=(16, 6))
ax.set(ylabel='Relative Importance');
ax.set(xlabel='Feature');
```
